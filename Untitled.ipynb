```python
import numpy as np
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import sklearn
from xgboost import XGBClassifier
from xgboost import plot_importance
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier,BaggingClassifier
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.metrics import classification_report,confusion_matrix,f1_score,roc_auc_score,roc_curve
from sklearn.model_selection import train_test_split
import xgboost as xgb
import lightgbm as lgb
import gc
gc.enable()
```

```python
#list of all dataframes
df_list=[df1,df2,df3,df4,df5,df6,df7,df8,df9,]
```

```python
def grouped_ID(dataframe):
    
    dataframe_group=dataframe.groupby('bookingID').median()
    #dataframe_group['number_of_trips']=dataframe.groupby('bookingID').count()['Accuracy']
    final=pd.merge(label,dataframe_group,on='bookingID')
    return final 

def feature(df1):

    df1['acc_product']=df1.acceleration_x * df1.acceleration_y * df1.acceleration_z
    
    df1['speed_acc_x']=df1.Speed * df1.acceleration_x
    df1['speed_acc_y']=df1.Speed * df1.acceleration_y
    df1['speed_acc_z']=df1.Speed * df1.acceleration_z

    df1['speed_gyro_x']=df1.Speed * df1.gyro_x
    df1['speed_gyro_y']=df1.Speed * df1.gyro_y
    df1['speed_gyro_z']=df1.Speed * df1.gyro_z

    return df1


def scale(df1):    
    
    scale=MinMaxScaler()
    scaled_df=scale.fit_transform(df1.drop(['bookingID','label'],1))
    scaled_df=pd.DataFrame(scaled_df,columns=df1.drop(['bookingID','label'],1).columns)

    scaled_df['label']=df1.label
    
    return scaled_df
```
